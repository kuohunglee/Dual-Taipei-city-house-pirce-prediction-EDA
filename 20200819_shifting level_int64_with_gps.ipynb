# 修改"shifting level"欄位

```python
# 41199筆的非重複地址 list
taipei['shifting level'].unique()
```

```python
# 去除非樓層項目
taipei['new_shifting level'] = taipei['shifting level'].str.replace('地下一層','',regex = True)
taipei['new_shifting level'] = taipei['new_shifting level'].str.replace('[^，一二三四五六七八九十層全]','',regex = True)
```

```python
taipei['new_shifting level'].unique()
```

```python
#取代用變數
num1 = ['一','二','三','四','五','六','七','八','九']
rnum1 = []
for i in range(1,10):
    l = str(i)
    rnum1.append(l)
num2 = ['2十層','3十層','4十層','5十層','6十層','7十層','8十層','9十層']
rnum2 = []
for i in range(2,10):
    l = str(i)+'0層'
    rnum2.append(l)
num3 = ['2十','3十','4十','5十','6十','7十','8十','9十']
rnum3 = []
for i in range(2,10):
    l = str(i)
    rnum3.append(l)

num4 = '十層'
rnum4 = '10層'    
    
num5 = '十'
rnum5 = '1'

num6 = '，'
rnum6 = ''

num7 = '層'
rnum7 = '，'

num8 = '，，'
rnum8 = '，'

num9 = '^，'
rnum9 = ''

num10 = '全'
rnum10 = '5'
```

```python
#帶入變數做置換
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num1,rnum1,regex=True)
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num2,rnum2,regex=True)
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num3,rnum3,regex=True)
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num4,rnum4,regex=True)
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num5,rnum5,regex=True)
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num6,rnum6,regex=True)
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num7,rnum7,regex=True)
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num8,rnum8,regex=True)
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num9,rnum9,regex=True)
```

```python
# 取代含一層的房子為'1'
level_1 = [r'^1，(.+)?',r'.+，1，(.+)?']
level_1_rep = [r'1',r'1']
taipei['new_shifting level'] = taipei['new_shifting level'].replace(level_1, level_1_rep, regex=True)
```

```python
#取代空值為'0'
taipei['new_shifting level'] = taipei['new_shifting level'].fillna('0')
```

```python
#取代空字串為'0'
taipei['new_shifting level'] = taipei['new_shifting level'].str.replace(r'^\s*$','0',regex=True)
```

```python
taipei['new_shifting level'].unique()
```

```python
#把2~9樓層數轉移的樓層歸類為第2類(不包含1樓)
taipei['new_shifting level'] = taipei['new_shifting level'].str.replace('^[2-9]，(.+)?','2',regex =True)
```

```python
taipei['new_shifting level'] = taipei['new_shifting level'].str.replace('1[0-9].+','3',regex =True)
```

```python
taipei['new_shifting level'] = taipei['new_shifting level'].str.replace('((2[0-9])|(3[0-9])|(4[0-9])).+','4',regex =True)
```

#把10樓層數以上轉移的樓層歸類為第3類
taipei['new_shifting level'] = taipei['new_shifting level'].str.replace('((1[0-9])|(2[0-9])|(3[0-9])|(4[0-9])).+','4',regex =True)

```python
taipei['new_shifting level'].unique()
```

```python
#驗證20層以上的數量清洗前後是否有誤差
a = taipei['new_shifting level'].str.contains('4').sum()
b = taipei['shifting level'].str.contains('[二三四]十[一二三四五六七八九]?層').sum()
c = taipei['shifting level'].str.contains('十九層，二十層').sum()
print(a) #分類完成後20層以上的房屋轉移數量
print(b) #原始資料20層以上的房屋轉移數量
print(c) #分類完成後涵蓋19&20層的房屋轉移數量(會因為19層被歸類在第三類)
```

```python
#把全層轉移歸類成第4類
taipei['new_shifting level'] = taipei['new_shifting level'].replace(num10,rnum10,regex=True)
taipei['new_shifting level'].unique() #再次驗證unique值是否有符合期待分類
```

```python
#把['new_shifting level']轉換成int
taipei['new_shifting level']=taipei['new_shifting level'].astype('int64')
```
