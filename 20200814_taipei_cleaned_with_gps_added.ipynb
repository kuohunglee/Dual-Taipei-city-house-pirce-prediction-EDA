# 匯入房屋(非重複)地址&地址GPS List

```python
# 41199筆的非重複地址 list
f=open("D:\\資策會期末專題資料\\房地產\\政府資料_雙北房地產\\經緯度地址\\dual_taipei_addr_list.json",'r',encoding='utf-8')
strjson = f.read()
f.close()
addr_list = json.loads(strjson)

# 41199筆的非重複地址的GPS list
f=open("D:\\資策會期末專題資料\\房地產\\政府資料_雙北房地產\\經緯度地址\\dual_taipei_geo_list_merge.json",'r',encoding='utf-8')
strjson = f.read()
f.close()
geo_list = json.loads(strjson)
print(len(addr_list))
print(len(geo_list))
```

#  把GPS中的經緯度list格式轉成字串的組合

```python

geo_str_list = []
for i in geo_list:
    
    if i == None:
        i='nan'
        geo_str_list.append(i)
    else:
        x=i[0]
        y=i[1]
        z = f'{x},{y}'
        geo_str_list.append(z)

geo_str_list
```

```python
# 確保地址每項值都為string
addr_str_list =[]
for i in addr_list:
    a = str(i)
    addr_str_list.append(a)

addr_str_list
```

```python
# 檢查list長度是否有跑掉
len(addr_str_list)
```

```python
# 檢查list長度是否有跑掉
len(geo_str_list)
```

```python
# 確保'new addr'每個欄位值都是string
df['addr'] = df['addr'].astype('str')
df['addr']
```

#  用地址和GPS的List來做dataframe['new addr']地址的置換

```python
df['WGS84'] = df['addr'].replace(addr_str_list,geo_str_list)
```
