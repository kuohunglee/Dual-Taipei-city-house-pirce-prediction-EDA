# 拆分transaction pen number欄位

```python
taipei_house['transaction pen number']=taipei_house['transaction pen number'].str.replace('建物',',建物')
taipei_house['transaction pen number']=taipei_house['transaction pen number'].str.replace('車位',',車位')
split_0=taipei_house['transaction pen number'].str.split(',').str.get(0)
split_1=taipei_house['transaction pen number'].str.split(',').str.get(1)
split_2=taipei_house['transaction pen number'].str.split(',').str.get(2)
x=list(split_0)
y=list(split_1)
z=list(split_2)
df_land = pd.DataFrame(x, columns=['Land'])
df_building = pd.DataFrame(y, columns=['Building'])
df_parking_space = pd.DataFrame(z, columns=['Parking Space'])

taipei_house['Land'] = df_land
taipei_house['Building'] = df_building
taipei_house['Parking Space'] = df_parking_space

taipei_house['Land']=taipei_house['Land'].str.replace('土地','')
taipei_house['Building']=taipei_house['Building'].str.replace('建物','')
taipei_house['Parking Space']=taipei_house['Parking Space'].str.replace('車位','')
taipei_house = taipei_house.drop(['transaction pen number'],axis=1)

taipei_house['Land'] = taipei_house['Land'].astype('int')
taipei_house['Building']=taipei_house['Building'].astype('int')
taipei_house['Parking Space']=taipei_house['Parking Space'].astype('int')
```

# 總層數欄位數字化(float64)

```python
#取代用變數
num1 = ['一','二','三','四','五','六','七','八','九']
rnum1 = []
for i in range(1,10):
    l = str(i)
    rnum1.append(l)
num2 = ['2十層','3十層','4十層','5十層','6十層','7十層','8十層','9十層']
rnum2 = []
for i in range(2,10):
    l = str(i)+'0層'
    rnum2.append(l)
num3 = ['2十','3十','4十','5十','6十','7十','8十','9十']
rnum3 = []
for i in range(2,10):
    l = str(i)
    rnum3.append(l)

num4 = '十層'
rnum4 = '10層'    
    
num5 = '十'
rnum5 = '1'

num6 = '地下層'
rnum6 = '-1層'

num7 = '地下'
rnum7 = '-'

num8 = '層'
rnum8 = ''
```

```python
#帶入變數做置換
taipei_house['total floor number'] = taipei_house['total floor number'].replace(num1,rnum1,regex=True)
taipei_house['total floor number'] = taipei_house['total floor number'].replace(num2,rnum2,regex=True)
taipei_house['total floor number'] = taipei_house['total floor number'].replace(num3,rnum3,regex=True)
taipei_house['total floor number'] = taipei_house['total floor number'].replace(num4,rnum4,regex=True)
taipei_house['total floor number'] = taipei_house['total floor number'].replace(num5,rnum5,regex=True)
taipei_house['total floor number'] = taipei_house['total floor number'].replace(num6,rnum6,regex=True)
taipei_house['total floor number'] = taipei_house['total floor number'].replace(num7,rnum7,regex=True)
taipei_house['total floor number'] = taipei_house['total floor number'].replace(num8,rnum8,regex=True)
```

```python
#轉換成float64
taipei_house['total floor number'] = taipei_house['total floor number'].astype('float')
```
